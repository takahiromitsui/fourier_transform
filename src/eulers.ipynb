```python
import matplotlib.pyplot as plt
from sympy import symbols, diff
import cmath
import math
import numpy as np
```

```python
# example of 2 to the x
def ecoli_growth(minutes:int):
    generation_time = 20
    return 2**(minutes/generation_time)


time = range(0,101)

growth = [ecoli_growth(t) for t in time]
fig = plt.figure(figsize=(4, 3))
plt.plot(time, growth)
plt.xlabel('Time (minutes)')
plt.ylabel('Number of cells')
plt.title('E. coli growth')

plt.show()
```

```python
def eulers_formula(devisor):
    theta = cmath.pi/devisor
    answer = cmath.exp(1j * theta)
    real_part = round(answer.real, 10)
    imag_part = round(answer.imag, 10)
    return complex(real_part, imag_part)
```

```python
# expected -1
answer = eulers_formula(1)
print(answer)
```

### Let's think about general e^x

```python
def exp_taylor(n:int):
    approx = 0
    for i in range(n):
        approx += 1/math.factorial(i)
    exact = cmath.exp(1)
    return approx, exact
```

```python
# Plot Taylor series and actual values

numbers = range(0,10)
results = [exp_taylor(i) for i in numbers]
# Extract the approximate and exact values from the results
approx_values = [result[0] for result in results]
exact_values = [result[1] for result in results]
plt.plot(numbers, approx_values, label='Approximate')
plt.plot(numbers, exact_values, label='Exact')

plt.legend()
plt.xlabel('n')
plt.ylabel('Value')
plt.title('Comparison of Taylor's)
plt.show()
```
